```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048
dtype = None
load_in_4bit = True

# "unsloth/Llama-3.2-1B-Instruct-bnb-4bit",
# "unsloth/Llama-3.2-1B-Instruct",
# "meta-llama/Llama-3.2-1B-Instruct",

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Llama-3.2-1B-Instruct",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit
)
```

### Function

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
def ask(question, max_new_tokens=1024, temperature=0.7, top_p=0.5):
    # 1. Format pesan ke template chat
    messages = [
        {"role": "user", "content": question}
    ]

    # 2. Tokenisasi input (ini mengembalikan dictionary)
    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt"
    ).to(model.device)

    # 3. Generate output
    # Gunakan **inputs untuk membongkar dictionary (input_ids & attention_mask)
    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        temperature=temperature,
        top_p=top_p,
    )

    # 4. Slicing yang benar
    # Ambil panjang sequence dari 'input_ids' di dalam dictionary inputs
    input_length = inputs['input_ids'].shape[1]

    # Decode hasil mulai dari setelah input_length
    response = tokenizer.decode(outputs[0][input_length:], skip_special_tokens=True)

    print(response)
```
